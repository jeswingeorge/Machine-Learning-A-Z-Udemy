```python
computers = pd.read_csv("datasets/computers.csv")
```

#### Validating the Linearity Assumption Using Scatter Plot

```python
fig, axes = plt.subplots(figsize=(12,8))
axes.scatter(x="Units",y="Minutes",data=computers)
axes.set_xlabel("Units")
axes.set_ylabel("Minutes")
axes.set_title("Plot of time taken to repair computers vs. number of units being replaced")
```

#### Validating the Linearity Assumption Using Residuals vs. Fitted Plot

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(X=computers.loc[:,["Units"]], y=computers.loc[:,["Minutes"]])
```

```python
predictions = model.predict(computers.loc[:,["Units"]])
predictions.shape
```

```python
sns.residplot(predictions.reshape(-1),'Minutes', data=computers,lowess=True,
                                  line_kws={'color': 'red', 'lw': 1, 'alpha': 1})
plt.xlabel("Fitted values")
plt.title('Residual plot')
```

```python
# example 2
x = np.arange(1,101)
y = x**2
model_1 = LinearRegression().fit(X= x.reshape(-1,1), y=y)
predicted_values = model_1.predict(x.reshape(-1,1))
print("Intercept:", model_1.intercept_,"\nCoefficients:", model_1.coef_[0])
```

```python
sns.residplot(predicted_values.reshape(-1),y,lowess=True,
                                  line_kws={'color': 'red', 'lw': 1, 'alpha': 1})
plt.xlabel("Fitted values")
plt.title('Residual plot')
```

#### Validating the Assumption That Errors are Normally Distributed

```python
residuals = computers["Minutes"] - predictions.reshape(-1)
```

```python
predictions = model.predict(computers.loc[:,["Units"]])
residuals = computers["Minutes"] - predictions.reshape(-1)
plt.figure(figsize=(7,7))
stats.probplot(residuals, dist="norm", plot=plt)
plt.title("Normal Q-Q Plot")
```

#### shapiro wilk test

```python
stats.shapiro(residuals)
```

#### Validating the Assumption of Homoscedasticity

```python
residuals = computers["Minutes"] - predictions.reshape(-1)
model_norm_residuals_abs_sqrt=np.sqrt(np.abs(residuals))
plt.figure(figsize=(7,7))
sns.regplot(predictions.reshape(-1), model_norm_residuals_abs_sqrt,
              scatter=True,
              lowess=True,
              line_kws={'color': 'red', 'lw': 1, 'alpha': 0.8})
plt.ylabel("Standarized residuals")
plt.xlabel("Fitted value")
```

#### Computing the Coefficient of Determination

```python
SST = sum((computers.Minutes.mean() - computers.Minutes)**2)
SST
```

```python
SSE = sum(residuals**2)
SSE
```

```python
SSR = SST - SSE
SSR
```

```python
Rsq = SSR/SST
Rsq
```

```python
Rsq1 = model.score(computers[["Units"]],computers["Minutes"])
Rsq1
```
